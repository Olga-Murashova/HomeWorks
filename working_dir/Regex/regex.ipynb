**Задание 1.** Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
import re
```

```python
def validate_car_sign(car_sign):
    pattern = r'([АВЕКМНОРСТУХ])(\d{3})([АВЕКМНОРСТУХ]{2})(\d{2,3})$'
    match = re.search(pattern, car_sign)
    result = print(f'Найден номер {match[2]}, найден регион {match[4]}')
    
    return result

validate_car_sign('А148МЕ199')
```

```python
email_regex = re.compile(r"""[a-z0-9!#$%&'*+/=?^_`{|}~-]+(?:\.[a-z0-9!#$%&'*+/=?^_`{|}~-]+)*@(?:[a-z0-9](?:[a-z0-9-]*[a-z0-9])?\.)+[a-z0-9](?:[a-z0-9-]*[a-z0-9])?""")

def mail_validate(mail):
    """ проверка по стандарту RFC 2822.
    Это регулярное выражение состоит из двух частей: часть до @, и часть после символа @. 
    Есть два варианта части перед @: она может состоять из ряда букв, цифр и некоторых символов, включая одну или несколько точек. 
    Тем не менее, точки не могут появляться последовательно, в начале или конце адреса электронной почты"""
    if email_regex.search(mail):
        return True
    else:
        return False

mail_validate('Kdavis@gmail.com')
```

**Задание 3.** Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
def remove_repeats(string):
    return re.sub(r'\b([^\W\d_]+)(\s+\1)+\b', r'\1', re.sub(r'\W+', ' ', string).strip())

remove_repeats('bla bla sh sh sh')
```

**Задание 4**. Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:   
+7(xxx)-xxx-xx-xx  
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера.
Примеры валидных форматов:
- +7 955 555-55-55;
- 9555555555;
- 8(955)555-55-55;
- +7 955 555 55 55;
- 7(955) 555-55-55;
- +7 955+555+55+55.

```python
def mobile_number(mob_number):
    number_pattern = re.compile(r'^(\+7|7|8)?[\s\-]?\(?([489][0-9]{2})\)?[\s\-|\+]?([0-9]{3})[\s\-|\+]?([0-9]{2})[\s\-|\+]?([0-9]{2}$)')
    if number_pattern.search(mob_number):
        num_list = number_pattern.split(mob_number)
        num_list = num_list[2:-1]
        result = '+7 ({}) {}-{}-{}'.format(*num_list)
        return result
    else:
        return 'Введен не корретный номер'
    
mobile_number('9555555555')
```
