Построить частотный график для двух параметров: Freedom и positive affect. Какое это распределение?

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.scatter(df['Positive affect'], df['Freedom'], c='blue', alpha=0.4)
ax.grid(True)
plt.xlabel('Positive affect')
plt.ylabel('Freedom')
plt.show()
```

Найти медиану, моду, математическое ожидание, стандартное отклонение, дисперсию, минимум и максимум, найти выбросы (если есть), размах для колонки Freedom

```python
import statistics
```

```python
print('Медиана: ', statistics.median(df['Freedom']))
# print('Мода: ', statistics.mode(df['Freedom'])) 
print('Мат.ожидание: ', statistics.mean(df['Freedom']))
print('Станд.отклон: ',statistics.pstdev(df['Freedom']))
print('Мин: ',min(df['Freedom']))
print('Макс: ',max(df['Freedom']))
print('Выбросы: ',df['Freedom'].isnull().sum())
print('Размах: ', (max(df['Freedom']) - min(df['Freedom'])))
```

Почему нельзя найти моду?

Посмотреть корреляцию между Freedom и positive affect (графически и математически). Графически - plt.scatter, математически - посчитать коэффициент Пирсона. О чем говорит знак коэффициента?

```python
x = df['Freedom']
y  = df['Positive affect']
plt.xlabel('выборка Freedom')
plt.ylabel('выборка Positive affect')
plt.grid(True)
plt.scatter(x, y, marker='.')
```

```python
from scipy.stats import pearsonr
```

```python
x = df['Freedom']
y  = df['Positive affect']

corr, p_value = pearsonr(x, y)
plt.scatter(x, y, marker='.')
plt.grid(True)
print('Коэффициент корреляции Пирсона:', corr)
```

**Комментарий:** Чем коэффициент ближе к 0, тем более независимы переменные, положительный знак коэффциента говорит о том, что корреляция сонаправлена.

Нормализовать данные колонки Freedom (централизация + масштабирование)

```python
import numpy as np
```

```python
#централизация
Freedom_c = df['Freedom']
print(statistics.mean(Freedom_c))
```

```python
#нормализация
Freedom_n = Freedom_c  - statistics.mean(Freedom_c)
print(statistics.mean(Freedom_n))
```

```python
#масштабирование
min(Freedom_n)
max(Freedom_n)
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))

Freedom_m= min_max_scaler.fit_transform(np.array(Freedom_n).reshape(-1, 1))
print(min(Freedom_m), max(Freedom_m))
```
